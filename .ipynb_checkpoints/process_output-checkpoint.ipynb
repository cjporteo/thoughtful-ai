```python
import os
import sqlite3
import random

def get_quote():
    conn = sqlite3.connect('quotes.db')
    c = conn.cursor()

    c.execute(
        """
        SELECT quote 
        FROM quotes
        WHERE LENGTH(quote) <= 140 AND english = 'Yes'
        """
    )

    quotes_in_db = set()

    for row in c:
        quotes_in_db.add(row[0])

    files = []
    for file in os.listdir('./output'):
        files.append(file)
    file = sorted(files, reverse=True)[0]

    print('Reading from: ' + file, end='\n\n')

    model_output = None

    with open('./output/' + file, 'r') as f:
        model_output = f.read()

    def filter_quote(quote):
        if quote in quotes_in_db or len(quote) > 140:
            return False

        blacklist_words = [
            'god',
            'lord'
            'jesus',
            'man',
            'woman'
            'suicide',
            'fuck',
            'shit',
            'rape'
        ]

        if any(blw in quote.lower() for blw in blacklist_words):
            return False

        return True


    quote_candidates = list(filter(filter_quote, model_output.split('<|endoftext|>')[1:-1]))

    for qc in quote_candidates:
        print(qc, end='\n\n')
    
    return random.choice(quote_candidates)
```
