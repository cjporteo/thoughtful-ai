```python
import sqlite3
import pandas as pd
```

```python
conn = sqlite3.connect('quotes.db')
c = conn.cursor()

df_authors = pd.read_sql_query(
    """
    SELECT author, SUM(likes) as total_likes, COUNT(1) as num_quotes,
        1.0*SUM(likes)/COUNT(1) as avg_likes
    FROM quotes
    GROUP BY author
    ORDER BY total_likes DESC
    """,
    conn
)

df_authors.head(50)
```

```python
c.execute(
    """
    SELECT quote 
    FROM quotes
    WHERE LENGTH(quote) <= 140 AND english = 'Yes'
    """
)

output_text = ''

for row in c:
    quote_text = row[0]
    if quote_text.count('"') % 2 == 0
        output_text += quote_text + '<|endoftext|>'

with open('../gpt-2-simple/quotes_training.txt', 'w', encoding='utf-8') as f:
    f.write(output_text)
```

```python
c.close()
conn.close()
```
